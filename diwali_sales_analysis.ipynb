# Exploratory Data Analysis

### Gender

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Gender', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum()
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(data = sales_gen, x = 'Gender', y = 'Amount', hue = 'Gender')
```

#### *Based on the graph, it can be concluded that the majority of buyers are female, and their purchasing power is also greater than that of men.*

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total amount vs age groups
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(data = sales_age, x = 'Age Group', y = 'Amount', hue = 'Age Group')
```

#### *The graphs indicate that most of the buyers are females aged between 26 and 35 years.*

## State

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(16,7)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders', hue = 'State')
```

```python
# total amount/sales for top 10 states
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(16,7)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount', hue = 'State')
```

#### *From the above graphs, it can be concluded that the majority of orders, as well as the highest total sales or amounts, come from Uttar Pradesh, Maharashtra, and Karnataka, respectively.*

## Marital Status

```python
sns.set(rc = {'figure.figsize':(7,5)})
ax = sns.countplot(data = df, x = 'Marital_Status', hue = 'Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc = {'figure.figsize':(6,5)})
sales_mar_gen = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(data = sales_mar_gen, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
```

#### *The graphs above show that the majority of buyers are married women with significant purchasing power.*

## Occupation

```python
sns.set(rc = {'figure.figsize':(20,7)})
ax = sns.countplot(data = df, x = 'Occupation', hue = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_job = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_job, x = 'Occupation', y = 'Amount', hue = 'Occupation')
```

#### *From above graphs, we can see that most of the buyers are working in IT, Healthcare and Aviation sector.*

## Product Category

```python
sns.set(rc = {'figure.figsize':(25,7)})
ax = sns.countplot(data = df, x = 'Product_Category', hue = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,7)})
sales_product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10) # top 10 categories

sns.barplot(data=sales_product, x='Product_Category', y='Amount', hue='Product_Category')
```

#### *From above graphs we can see that most of the sold products are from Food, Clothing & Apparel and Electronics & Gadgets category.*

### Product ID

```python
# top 10 most sold products based on product_ID
sales_product = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data=sales_product, x='Product_ID', y='Orders', hue='Product_ID')
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
